# 2. 클러스터링으로 회원들 그룹화

# 2.1 데이터 정리

```python
# end_date 탈퇴일을 제외하곤 결측치가 없다. 
# 고객데이터를 이용하여 그룹화합니다. 미리 정해진 정답이 없기에 비지도학습 클러스터링을 이용하겠습니다. 

# 고객의 한 달 이용 이력에 대한 mean, median, max, min, membership_period 를 불러와 변수로, 그룹을 나눕니다. 
customer_clustering = customer[["mean", "median", "max", "min", "membership_period"]]
customer_clustering.head()
```

```python
# 컬럼명 헷갈리니까 알아보기 쉽게 수정합니다. 

customer_clustering.columns = [
    "월평균값","월중앙값","월최댓값","월최솟값","회원기간"
]
customer_clustering.head()
```

# 2.2 표준화 스케일링

```python
# 위의 데이터를 보면, 각 컬럼별로 수치 값이 다릅니다. 
# mean의 경우 한 자리 수이지만, membership_perio는 두자리 수 입니다. 
# 표준화 작업 필요

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
customer_clustering_sc = sc.fit_transform(customer_clustering)
```

# 2.3 군집화 k-means

```python
# 클러스터링 그룹화 진행 
# k-means : 가장 전통적인 방법으로서, 변수간의 거리를 기반으로 그룹화함 
# 그룹화 하는 k의 수는 4개로 지정하겠습니다. 

from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=0)      # k의 수 지정하여 KMeans 준비 
clusters= kmeans.fit(customer_clustering_sc)       # 표준화한 customer_clustering_sc에 Kmeans 적용 
customer_clustering["cluster"] = clusters.labels_  # 적용값을 기존데이터의 새로운 컬럼에 추가 
print(customer_clustering["cluster"].unique())     # 클러스터링 넘버 출력 (k가 4개이니까 0,1,2,3)

customer_clustering.head()
```

# 3. 결과분석

```python
# 보유한 클러스터링 넘버별로 고객 수 카운트
customer_clustering.groupby("cluster").count()

# 여기선 3번 그룹의 회원수가 1342명으로 제일 많습니다. 
```

```python
# 각 그룹별 특징 파악을 위해, 평균값도 계산해봅니다. 
customer_clustering.groupby("cluster").mean()

# 1번그룹의 회원수는 846명이지만, 월평균이용은 다소 높습니다. 
# 2번 그룹은 다른 그룹보다 회원기간이 매우 깁니다. 
```

# 4. 결과 시각화

# 4.1 차원축소(주성분분석PCA)

```python
# 클러스터링에 사용된 변수는 5개이며, 5차원입니다. 
# 그러나 x와 y축으로 평평한 2차원의 그림으로 나타내고 싶습니다. 
# 5차원 -> 2차원을 표현하기 위해 차원축소를 해야합니다. 

# 대표적으로 주성분분석으로 표준화합니다. 
from sklearn.decomposition import PCA

X = customer_clustering_sc                       # 스케일링 및 kmeans 적용값을 X에 저장
pca = PCA(n_components = 2)                      # 5개 고차원 데이터 -> 2차원으로 축소 / PCA 주성분분석 함수 이용 
pca.fit(X)                                       # 스케일링 및 kmeans 적용값을 PCA모델을 학습시킴
x_pca = pca.transform(X)                         # 학습값을 2차원으로 축소
pca_df = pd.DataFrame(x_pca)                     # 축소값을 데이터프레임으로 저장 
pca_df["cluster"] = customer_clustering["cluster"] # 저장값에 클러스터를 나타내는 컬럼값을 반영  

# 2차원으로 축소화한 값과 클러스터링 넘버 
pca_df.head()           
```

# 4.2 산점도 시각화 

```python
import matplotlib.pyplot as plt
%matplotlib inline 

# 기존데이터의 클러스터링 번호를 하나씩 하나씩 추출 
for i in customer_clustering["cluster"].unique():    
    # 클러스터링 번호 0 -> 1 -> 2 -> 3 이렇게 추출하는 행만 필터링해서 각각 tmp에 저장 
    tmp = pca_df.loc[pca_df["cluster"] == i]
    # 산점도로 표현 
    # tmp[0] 은 첫번째 주성분 PC1의 값이고, tmp[1]은 두번째 주성분 PC2의 값 
    plt.scatter(tmp[0], tmp[1])
```

# 6.2 지난달 데이터 추출(반복문)

```python
# 기간 확인 
list(uselog_months["연월"].unique())
```

```python
# 약 6개월동안 데이터를 활용해서 미래시점의 데이터를 예측할 수 있도록, 데이터 준비  
# 2018년4월 + 6개월 = 2018년 10월부터 시작해서 1~6개월전 값 반복문으로 뽑아내고, 그 데이터를 활용해서 미래 예측 
# 
#   |--------------------------------------------------------|
#   | 현재월   ||  1달전   |   2달전   |   3달전  | .. |   6달전   |
#   |--------------------------------------------------------|
# 1 | 201810 || 201809값 | 201808값 | 201807값 | .. | 201804값 | 1달전 ~6달전    |
# 2 | 201811 || 201810값 | 201809값 | 201808값 | .. | 201805값 | 데이터를 기준으로 |
# 3 | 201812 || 201811값 | 201810값 | 201809값 | .. | 201806값 | 현재월의 값을    |
# 4 | 201901 || 201812값 | 201811값 | 201810값 | .. | 201807값 | 추출하고        |
# 5 | 201902 || 201901값 | 201812값 | 201811값 | .. | 201808값 | 맞는지         |
# 6 | 201903 || 201902값 | 201901값 | 201812값 | .. | 201809값 | 확인한다.       |
#   |--------------------------------------------------------|
#     for i문 || 여기부터는 for j 문 --------------------------->|
#     tmp저장값||  tmp_before 저장값 --------------------------->|
```

```python
# 기간목록을 컬럼으로 저장   
year_months = list(uselog_months["연월"].unique()) 

# 위의 데이터 프레임으로 만들어야하니, 일단 빈 값의 프레임 생성
predict_data = pd.DataFrame()
```

```python
# 하나씩 집어넣을 반복문 만들기 ("201804"부터가 아닌 7번째 "201810"부터 시작, 해당 리스트 끝까지 ) 1 
for i in range(6, len(year_months)): 
    
    # 현재 201810 ~ 201903 까지 추출되는 반복문의 데이터를 하나하나 꺼낸 것만 uselog_months 데이터에 맞춰서 tmp 저장 
    tmp = uselog_months.loc[uselog_months["연월"]==year_months[i]]
    
    # 기존데이터중 201810 ~ 201903 만 추출한 값에서, 1개월 전부터 6개월 전까지 데이터만 추출 (0은 현재) 
    for j in range(1,7): 
        # 1~7개월전 데이터 하나씩 추출해서 기존데이터와 맞춘다. 이후 저장한다. 
        tmp_before = uselog_months.loc[uselog_months["연월"]==year_months[i-j]]

    # 위의 빈 데이터프레임에 저장한다. 
    predict_data = pd.concat([predict_data, tmp], ignore_index=True)
    
predict_data.head()

# 그럴싸 하지만 1달전~6달전 각각의 데이터값을 저장하여 붙히는게 없다. 
```

```python
# 하나씩 집어넣을 반복문 만들기 ("201804"부터가 아닌 7번째 "201810"부터 시작, 해당 리스트 끝까지 ) 2 
for i in range(6, len(year_months)): 
    
    # 현재 201810 ~ 201903 까지 추출되는 반복문의 데이터를 하나하나 꺼낸 것만 uselog_months 데이터에 맞춰서 tmp 저장 
    tmp = uselog_months.loc[uselog_months["연월"]==year_months[i]]
    
    # 기존데이터중 201810 ~ 201903 만 추출한 값에서, 1개월 전부터 6개월 전까지 데이터만 추출 (0은 현재) 
    for j in range(1,7): 
        # 1~7개월전 데이터 하나씩 추출해서 기존데이터와 맞춘다. 이후 저장한다. 
        tmp_before = uselog_months.loc[uselog_months["연월"]==year_months[i-j]]
        
#       ★★★★★ 없으면 "연월"열이 2개이므로 중복열이 발생하여 error  !!!!!!!!!!!! 
        del tmp_before["연월"]    # 이전 월 데이터의 "연월" 열을 삭제 

#       ★★★★★ 없으면 "count"열이 여러개 생성되므로 중복열 발생으로 error !!!!!!!
        tmp_before.rename(columns={"count":"count_{}".format(j-1)}, inplace=True)
        
        # 그리곤 tmp에 붙힌다. 
        tmp = pd.merge(tmp, tmp_before, on = "customer_id", how="left")

    # 위의 빈 데이터프레임에 저장한다. 
    predict_data = pd.concat([predict_data, tmp], ignore_index=True)
    
predict_data.head()
```

```python
# 하나씩 집어넣을 반복문 만들기 ("201804"부터가 아닌 7번째 "201810"부터 시작, 해당 리스트 끝까지 ) 3 
for i in range(6, len(year_months)): 
    
    # 현재 201810 ~ 201903 까지 추출되는 반복문의 데이터를 하나하나 꺼낸 것만 uselog_months 데이터에 맞춰서 tmp 저장 
    tmp = uselog_months.loc[uselog_months["연월"]==year_months[i]]

#   ★★★★★ 없으면 "count"열이 2개이므로 중복열이 발생하여 error !!!!!!!!!     
    tmp.rename(columns={"count":"count_pred"}, inplace=True)
    
    # 기존데이터중 201810 ~ 201903 만 추출한 값에서, 1개월 전부터 6개월 전까지 데이터만 추출 (0은 현재) 
    for j in range(1,7): 
        
        # 1~7개월전 데이터 하나씩 추출해서 기존데이터와 맞춘다. 이후 저장한다. 
        tmp_before = uselog_months.loc[uselog_months["연월"]==year_months[i-j]]
        
#       ★★★★★ 없으면 "연월"열이 2개이므로 중복열이 발생하여 error  !!!!!!!!!!!! 
        del tmp_before["연월"]    # 이전 월 데이터의 "연월" 열을 삭제 
        
#       ★★★★★ 없으면 "count"열이 여러개 생성되므로 중복열 발생으로 error !!!!!!!
        tmp_before.rename(columns={"count":"count_{}".format(j-1)}, inplace=True)
        
        # 그리곤 tmp에 붙힌다. 
        tmp = pd.merge(tmp, tmp_before, on = "customer_id", how="left")
        
    # 위의 빈 데이터프레임에 저장한다. 
    predict_data = pd.concat([predict_data, tmp], ignore_index=True)
    
predict_data.head()

# 참고로 여기서 생기는 NaN값은 가입기간이 6개월 미만일경우 
# count_pred는 현재연월의 이용횟수 입니다. 
# count_0은 1달전
# count_1은 2달전 
# count_5는 6달전 
```

```python
# 이제 이 데이터를 활용해서 과거데이터(count_0 ~ count5)를 이용하여 현재연월을 예측할 수 있다. 
# count_pred (정답지) 와 비교해서 얼마나 정확도가 있는지 비교할수도 있다.

print(predict_data.isnull().sum())

predict_data.info()
```

```python
# NaN값 결측치 있는 데이터들 모두 제거 == 6개월이상 이용중인 회원들만 있음 
predict_data = predict_data.dropna()

# 다시 데이터프레임 순서 리셋 
predict_data = predict_data.reset_index(drop=True)

print(predict_data.isnull().sum())

predict_data.info()
```

# 6.3.2 이용기간(m) 변수 추가

```python
# "연월" yyyymm 을 날짜형으로 바꾸어서, yyyy-mm-dd 변수 추가 
predict_data["now_date"] = pd.to_datetime(predict_data["연월"], format="%Y%m")
predict_data.head()
```

```python
# 날짜형으로 재변환 
predict_data["start_date"] = pd.to_datetime(predict_data["start_date"])
predict_data.head()
```

```python
# 날짜 계산하는 relativedelta 함수 
from dateutil.relativedelta import relativedelta 

# 날짜간 계산한 값을 넣어줄 컬럼 미리 생성 
predict_data["period"] = None

# 반복문으로 하나씩 row 꺼내서 now_date (현재) 빼기 start_date (가입일)
for i in range(len(predict_data)): 
    # 날짜계산 함수로 하나씩 대입해서 delta에 저장 
    delta = relativedelta(
        predict_data["now_date"][i], 
        predict_data["start_date"][i]
    )
    # delta에 저장된 날짜간 차이 (날짜형)을 years단위라면 12곱해야 Months 기준으로 됨 // 그리고나서 period 컬럼에 저장 
    predict_data["period"][i] = delta.years * 12 + delta.months

predict_data.head()
```

```python
# 아까 예측을 위한 과거데이터 만들때 2018년 4월부터 였으니, 가입일도 이때부터로 합니다. 
# list(uselog_months["연월"].unique())   # --> 201804 부터 201903 까지임 

predict_data = predict_data.loc[predict_data["start_date"] >= pd.to_datetime("20180401")]

# 확인 
print(predict_data["start_date"].min())
```

```python
#################### 머신러닝 모델 함수 불러오기
from sklearn import linear_model

# 모델 선정
model = linear_model.LinearRegression()

# 예측에 사용할 학습 데이터 
X = predict_data[["count_0", "count_1", "count_2", "count_3", "count_4", "count_5", "period"]]

# 예측하고 싶은 값 
y = predict_data["count_pred"]

#################### 데이터 분할 함수 불러오기 
import sklearn.model_selection

# X와 y를 각각 학습용 , 평가용으로 분리 (지정하지 않을경우 학습75%, 테스트용 25%) 
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X,y)

# 학습데이터를 이용해서 모델 구축 
model.fit(X_train, y_train)
```

```python
# 정확도 검증 
print(model.score(X_train, y_train)) # 0.607 --> 정확도 약 60% 
print(model.score(X_test, y_test))   # 0.611 --> 정확도 약 60% 
```

```python
# 선형회귀모델에서는 중요한게 coef 회귀계수도 확인해봐야 합니다. 
# 선형회귀모델에서 각 독립변수(X)가 y값에 얼마나 영향력을 미치는지 볼 수 있습니다. 

coef = pd.DataFrame(
    {
        "feature_names": X.columns, 
        "coefficient" : model.coef_
    }
)
coef

# count_0 바로 지난달을 나타내는 값이 회귀계수가 제일 높군요. 
```
