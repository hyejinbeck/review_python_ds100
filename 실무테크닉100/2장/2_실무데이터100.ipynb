# 2.1 결측치 확인 후 제거

```python
# 결측치 확인  (True면 NaN값이 있는거다)
# uriage_data.isnull()             # --> 데이터프레임 형태로 

uriage_data.isnull().any(axis=0)   # 각 column 별로 NaN값이 있는지 
```

```python
# NaN값이 있는 "item_price" 을 새로운 변수에 일단 저장 
# NaN값이 있으면 True 아니라면 False 값으로 나옵니다. 
flg_is_null = uriage_data["item_price"].isnull()

# NaN값이 있는 항목중 "item_name"을 하나씩 추출 == 그러니까 어떤 상품명이 NaN값인지 하나씩 뽑아야지 
for name in list(uriage_data.loc[flg_is_null, "item_name"].unique()):
# ----------------------
# item_name | item_price
# ----------------------
#   상품A    |    NaN         --> NaN값인 상품명만 추출 
#   상품B    |    100.0 
    
    # 결측값이 아닌 행중에 (~flg_is_null) "item_price"의 최대값을 가져옵니다. 
    price = uriage_data.loc[(~flg_is_null) & (uriage_data["item_name"] == name), "item_price"].max()
    # ----------------------
    # item_name | item_price
    # ----------------------
    #   상품A    |    NaN        
    #   상품A    |   1200       --> 그 중 최대값인 1200 을 가져오기 
    #   상품A    |    900
    
    # 결측값중엣 item_name이 trg인 항목의 가격을 "최대값"으로 채웁니다. 
    uriage_data["item_price"].loc[(flg_is_null) & (uriage_data["item_name"]==name)] = price
    # ----------------------
    # item_name | item_price
    # ----------------------
    #   상품A    |   1200        
    #   상품A    |   1200       --> 그 중 최대값인 1200 을 가져오기 
    #   상품A    |    900
    
uriage_data.head()
```

# 5. 데이터 집계

```python
# 구매년월을 기준으로 피벗테이블로 만들어서 상품별(A~Z) 각각 "몇 개씩" 판매되었는지 집계

new_data.pivot_table(index="구매년월", columns="상품명", aggfunc="size", fill_value = 0 )
#                    ------- 기준으로 묶기      ---- 상품명 별로 ----- 갯수 계산       ----- NaN값일 경우 0으로 채우기
```

```python
# 구매년월을 기준으로 피벗테이블로 만들어서 상품별(A~Z) 각각 "매출금액" 의 합계" 집계 

new_data.pivot_table(index="구매년월", columns="상품명", values = "상품가격", aggfunc="sum", fill_value = 0 )
#                     ------- 기준        ------ 상품명별로       ------ 금액의 합계 계산 ----     
```

```python
# 구매년월을 기준으로 피벗테이블로 만들어서 고객마다 몇개씩 구매했는지 
new_data.pivot_table(index="구매년월", columns="고객명", aggfunc="size", fill_value = 0 )
```

```python
# 구매년월을 기준으로 피벗테이블로 만들어서 고객마다 총 얼마를 구매했는지 합계 
new_data.pivot_table(index="구매년월", columns="고객명", values = "상품가격", aggfunc="sum", fill_value = 0 )
```

```python
# 구매년월을 기준으로 피벗테이블로 만들어서 각 지역별로 집계 
new_data.pivot_table(index="구매년월", columns="지역", aggfunc="size", fill_value=0)
```
