# 8강 : 수치 시뮬레이션을 이용하여, 고객들의 입소문으로 얼마나 매출이 오를지 예측 가능한가요? 

재구매 고객 20명의 SNS 연결 상태가 저장된 데이터를 통해, 입소문에 의한 정보 전파 과정을 확인해봅니다.

```python
# 서로간 SNS 연결상태를 나타내는 고객관계 데이터
# links.csv 재구매 고객 20명
# links_members.csv 재구매고객 540명

# info_members.csv 재구매고객 540명의 월별 이용현황 
```

# 1. 데이터 읽기

```python
import pandas as pd

df_links = pd.read_csv("links.csv")
df_links

# 연결되어 있으면 1, 없으면 0 
```

# 2. 인간관계 네트워크 가시화

```python
import networkx as nx
import matplotlib.pyplot as plt

# 그래프 객체 생성
G = nx.Graph()

# 노드 설정
NUM = len(df_links.index)
for i in range(1,NUM+1):
    node_no = df_links.columns[i].strip("Node")
    G.add_node(str(node_no))

# 엣지 설정
for i in range(NUM):
    for j in range(NUM):
        node_name = "Node" + str(j)
        if df_links[node_name].iloc[i]==1:
            G.add_edge(str(i),str(j))
        
# 그리기
# draw가 아닌 draw_networkx를 사용함으로서, 다른것과 연결이 많은 노드를 가운데로 위치 조정 
nx.draw_networkx(G,node_color="k", edge_color="k", font_color="w")
plt.show()

# 실행할 때마다 모양이 달라진다! 
```

# 3. 입소문에 의해 정보 전파되는 모습 가시화

10명 중 1명에게 전파한다고 가정정합니다 (10% 확률)

```python
import numpy as np

# 입소문의 전파 여부를 확률적으로 결정하는 함수 
def determine_link(percent):
    rand_val = np.random.rand()
    if rand_val<=percent:
        return 1
    else:
        return 0
```

```python
# 입소문 시뮬레이션 
def simulate_percolation(num, list_active, percent_percolation):
#                        --사람수 ------ 전달여부  ------- 확률 
    for i in range(num):
        if list_active[i]==1:
            for j in range(num):
                node_name = "Node" + str(j)
                if df_links[node_name].iloc[i]==1:
                    if determine_link(percent_percolation)==1:
                        list_active[j] = 1
    return list_active
```

```python
percent_percolation = 0.1      # 전파 확률 
T_NUM = 48                     # 입소문 전파되는 기간 (개월 수) 
NUM = len(df_links.index)      # 사람 수
```

```python
list_active = np.zeros(NUM)
list_active[0] = 1

list_timeSeries = []
for t in range(T_NUM):
    list_active = simulate_percolation(NUM, list_active, percent_percolation)
    list_timeSeries.append(list_active.copy())
```

```python
# 액티브 노드 가시화 #
def active_node_coloring(list_active):
    #print(list_timeSeries[t])
    list_color = []
    for i in range(len(list_timeSeries[t])):
        if list_timeSeries[t][i]==1:
            list_color.append("r")
        else:
            list_color.append("k")
    #print(len(list_color))
    return list_color
```

```python
# 0개월 일 때 모습 
t = 0
nx.draw_networkx(G,font_color="w",node_color=active_node_coloring(list_timeSeries[t]))
plt.show()
```

```python
# 11개월 뒤 일 때 모습 
t = 11
nx.draw_networkx(G,font_color="w",node_color=active_node_coloring(list_timeSeries[t]))
plt.show()
```

```python
# 11개월 뒤 일 때 모습 
t = 35
nx.draw_networkx(G,font_color="w",node_color=active_node_coloring(list_timeSeries[t]))
plt.show()
```

# 4. 입소문에 의해 정보 전파되는 결과 시계열 그래프화

```python
# 시계열 그래프 그리기
list_timeSeries_num = []
for i in range(len(list_timeSeries)):
    list_timeSeries_num.append(sum(list_timeSeries[i]))

plt.plot(list_timeSeries_num)
plt.show()
```
