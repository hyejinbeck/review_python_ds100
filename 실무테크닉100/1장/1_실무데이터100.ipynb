### 데이터 가공 및 통계 분석

```python
# 상품 데이터 
import pandas as pd 

# 고객정보 데이터 
customer_master = pd.read_csv('1_customer_master.csv') 
customer_master
```

```python
# 상품 데이터 
item_master = pd.read_csv('1_item_master.csv')
item_master
```

```python
# 구매내역 데이터 
transaction_1 = pd.read_csv('1_transaction_1.csv')
transaction_1
```

```python
# 구매내역 데이터 
transaction_2 = pd.read_csv('1_transaction_2.csv')
transaction_2
```

```python
# 구매내역 상세 데이터 
transaction_detail_1 = pd.read_csv('1_transaction_detail_1.csv')
transaction_detail_1
```

```python
# 구매내역 상세 데이터 
transaction_detail_2 = pd.read_csv('1_transaction_detail_2.csv')
transaction_detail_2
```

```python
transaction = pd.concat([  transaction_1, transaction_2 ] , ignore_index=True)   # 5000 rows + 1786 rows = 6786 
transaction
```

```python
transaction_detail = pd.concat([  transaction_detail_1, transaction_detail_2 ] , ignore_index=True)   # 5000 rows + 2144 rows = 7144
transaction_detail
```

```python
# 새변수만들기 =  merge (    기존변수명             붙힐변수명의    [[   붙힐컬럼명1 ,    붙힐컬럼명2    , 붙힐컬럼명3..]],  공통컬럼명 , 왼쪽기준 오른쪽에 붙힐거임
join_data = pd.merge( transaction_detail , transaction[["transaction_id", "payment_date","customer_id"]], on="transaction_id", how="left")
join_data
```

```python
print(len(transaction_detail))    # 기존변수 
print(len(transaction))           # 에 붙힐 변수 
print(len(join_data))             # 기존변수에 가로로 컬럼이 붙었으니 rows 변동사항 없음 check 
```

```python
#  합칠변수명 =         기존변수명   ,    추가변수명       ,      공통컬럼명      , 왼쪽기준 오른쪽에 붙히기
join_data = pd.merge(join_data ,  customer_master , on = "customer_id" , how = "left") 
join_data = pd.merge(join_data ,  item_master     , on = "item_id"     , how = "left") 
join_data.head()
```

```python
join_data["price"] = join_data["quantity"] * join_data["item_price"]
join_data.head()
```

```python
#    payment_date 를 날짜형 datetime 으로 바꿔줍니다.           
join_data["payment_date"] = pd.to_datetime(join_data["payment_date"])       # 2019-02-01 01:36:57

#    새 컬럼으로 만들어줍니다.     
join_data["payment_month"] = join_data["payment_date"].dt.strftime("%Y%m")  # 201902 

#    두 컬럼만 비교해줍니다. 
join_data[["payment_date", "payment_month"]].head()
```

```python
#            월별컬럼기준으로 묶어서      매출의 합계를 볼거야. 
join_data.groupby("payment_month")["price"].sum()
```

```python
join_data["item_name"].unique()
```

```python
#            월별컬럼과 아이템명 기준으로 묶어서       매출과 갯수의 합계를 볼거야
join_data.groupby(["payment_month", "item_name"])[["price", "quantity"]].sum()
```

```python
pd.pivot_table(                        # 가로정렬할수 있게 피벗테이블로 만들기 
    join_data,                         # 데이터 
    index = 'item_name',               # 세로    (아이템명)  그룹기준1
    columns = 'payment_month',         # 가로    (월별)     그룹기준2
    values = ['price','quantity'],     # 값     (합계) 
    aggfunc = 'sum'                    # 값 기준 (합계)
)
```

```python
graph_data = pd.pivot_table(                        # 가로정렬할수 있게 피벗테이블로 만들기 
            join_data,                         # 데이터 
            index = 'item_name',               # 세로    (아이템명)  그룹기준1
            columns = 'payment_month',         # 가로    (월별)     그룹기준2
            values = ['price','quantity'],     # 값     (합계) 
            aggfunc = 'sum'                    # 값 기준 (합계)
            )
graph_data.head()
```
